```python
finalData = pd.read_csv('/project/hackathon/hackers06/shared/finalDatafromAkshay/FinalData.csv')
```

```python
finalData
```

```python
def getCat(x):
    if x > 0 and x <= 7:
        return 'LOW'
    elif x > 7 and x <= 21:
        return 'MEDIUM'
    elif x > 21 and x <= 60:
        return 'HIGH'
    else:
        return 'VERY HIGH'
finalData['y_actual'] = finalData['ONSET_DAYS'].apply(lambda x: getCat(x))
```
